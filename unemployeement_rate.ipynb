# max unemployment 

```python
df.groupby("Region").max("Unemployment Rate (%)")
```

```python
max_unemployment_by_region = df.groupby('Region')['Estimated Unemployment Rate (%)'].max()
plt.figure(figsize=(10, 6))
max_unemployment_by_region.plot(kind='bar', color='skyblue')
plt.title('Maximum Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Maximum Unemployment Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# min employeement rate per region

```python
df.groupby("Region").min("Unemployment Rate (%)")
```

```python
min_unemployment_by_region=df.groupby("Region").min("Unemployment Rate (%)")
plt.figure(figsize=(10, 6))
min_unemployment_by_region.plot(kind='bar', color='skyblue')
plt.title('Minimum Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Minimum Unemployment Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
